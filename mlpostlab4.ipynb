```python
from ucimlrepo import fetch_ucirepo

# Fetch dataset
heart_disease = fetch_ucirepo(id=45)
# Data (as pandas DataFrames)
x1 = heart_disease.data.features
y1 = heart_disease.data.targets

df = pd.concat([x1, y1], axis=1)
df
```

```python
df.shape
```

```python
df.describe()
```

```python
x=df.iloc[:,:-1].values  #input features
y=df.iloc[:,-1].values #output label
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#normalize the dataset
scaler = StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# Convert x_train to DataFrame to drop rows with NaNs
x_train_df = pd.DataFrame(x_train)

# Create a mask for rows without NaNs
mask = ~x_train_df.isna().any(axis=1)

# Apply the mask to both x_train and y_train
x_train = x_train[mask]
y_train = y_train[mask]

# Train the model
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
from sklearn.impute import SimpleImputer

# Define the imputer (same strategy as before)
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training data
x_train_imputed = imputer.fit_transform(x_train)

# Now you can use the imputer to transform x_test
x_test_imputed = imputer.transform(x_test)

# Test the model
y_pred = model.predict(x_test_imputed)
print(y_pred)
```

```python
#evaluation parameters
print("predicted: ",y_pred)
print("original:  ",y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy = accuracy_score(y_pred,y_test)
print(accuracy)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
